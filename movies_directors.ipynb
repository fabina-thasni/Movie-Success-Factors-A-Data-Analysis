## Introduction

## Dataset Overview

The merged dataset contains the following columns:

* movie_id: Unique ID for each movie.
* budget: Budget of the movie.
* popularity: Popularity score of the movie.
* revenue: Revenue generated by the movie.
* title: Title of the movie.
* vote_average: Average user rating (vote) of the movie.
* vote_count: Number of user votes.
* director_id: Unique ID of the director.
* year: Release year of the movie.
* month: Release month of the movie.
* day: Release day of the movie.
* director_name: Name of the director.
* gender: Gender of the director.
* net_profit: Profit made by the movie (Revenue - Budget).
* profit_margin: Profit margin of the movie.

# 1. Relationship Between Movie Budget and Revenue

```python
# Scatter plot to show the relationship between budget and revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(x='budget', y='revenue', data=df, hue='gender', palette='Set2')
plt.title('Movie Budget vs Revenue',color="red")
plt.xlabel('Budget (in millions)')
plt.ylabel('Revenue (in millions)')
plt.show()
```

* Higher movie budgets tend to generate higher revenues.
* Some low-budget films achieve significant revenue.
* Gender does not show a clear impact on the budget-revenue relationship.
* A few outliers with very high revenues exist regardless of budget.

# 2. Analyzing Net Profit by Director's Gender

```python
# Boxplot to show the distribution of net profit based on director's gender
plt.figure(figsize=(10, 4))
sns.boxplot(x='gender', y='net_profit', data=df, palette='Set1')
plt.title('Net Profit by Director\'s Gender')
plt.xlabel('Gender')
plt.ylabel('Net Profit (in millions)')
plt.show()
```

* Male directors have a wider range of net profits, with several extreme outliers in the high-profit range.
* Female directors have fewer outliers, suggesting less variability in net profits compared to male directors.
* The median net profit is similar for both male and female directors.

# 3.Correlation Heatmap of Movie Metrics

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
corr_matrix = df[['budget', 'revenue', 'popularity', 'vote_average', 'vote_count', 'net_profit']].corr()
sns.heatmap(corr_matrix, annot=True,  fmt='.2f')
plt.title('Correlation Heatmap of Movie Metrics')
plt.show()
```

* Revenue and Budget: Strong positive correlation (0.71). Higher budget often leads to higher revenue.
* Popularity and Revenue: Moderate positive correlation (0.59). More popular movies tend to earn more.
* Vote Count and Popularity: Strong positive correlation (0.78). More votes usually mean higher popularity.
* Net Profit and Revenue: Very strong positive correlation (0.98). Higher revenue almost always means higher net profit.
* Vote Count and Revenue: Significant correlation (0.75). Movies with more votes tend to earn more.
* Net Profit and Vote Count: Strong positive correlation (0.73). Higher net profit is often associated with more votes.

This heatmap helps understand how different financial and rating aspects of movies are interrelated, which can be valuable for analysts in the film industry.

# 4.Relationship Between Popularity and Vote Count Based on Director's Gender

```python
# Joint plot for popularity and vote count
plt.figure(figsize=(8, 6))
sns.jointplot(x='popularity', y='vote_count', data=df, kind='scatter', hue='gender', palette='muted')
plt.title('Popularity vs Vote Count')
plt.xlabel('Popularity')
plt.ylabel('Vote Count')
plt.tight_layout()
plt.show()
```

* Higher popularity leads to more votes for movies.
* Most movies have low to moderate popularity and vote count.
* Male-directed movies are more frequent in the dataset.
* A few outliers show very high popularity and vote counts.

# 5.Net Profit vs Budget by Director's Gender

```python
# Line plot for Net Profit vs Budget
plt.figure(figsize=(8, 6))
sns.lineplot(x='budget', y='net_profit', data=df, hue='gender', palette='Set1')
plt.title('Net Profit vs Budget')
plt.xlabel('Budget (in millions)')
plt.ylabel('Net Profit (in millions)')
plt.show()
```

* As the budget increases, the potential for higher net profit also increases.
* Male-directed movies show more volatility in net profit, especially with higher budgets.
* Female-directed movies tend to follow a more stable trend with less fluctuation in profit across varying budgets.

# 6.Distribution of Movie Budgets

```python
# Histogram for movie budgets
plt.figure(figsize=(8, 6))
sns.histplot(df['budget'], bins=30, kde=True, color='purple')
plt.title('Distribution of Movie Budgets')
plt.xlabel('Budget (in millions)')
plt.ylabel('Frequency')
plt.show()
```

* The majority of movies have lower budgets, mostly under $50 million.
  
* Very few movies have budgets exceeding $200 million, indicating that high-budget films are rare.

# 7.Movie Release Month vs Revenue

```python
# Barplot showing the average revenue based on the month of release
plt.figure(figsize=(10, 6))
sns.barplot(x=df['month'], y=df['revenue'], palette='coolwarm')
plt.title('Average Revenue by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Average Revenue (in millions)')
plt.show()
```

* Highest revenue months: May, june, and November.
* Lowest revenue months: January, September, and Feb.

# 8.Average Revenue by Release Day

```python
# Barplot showing the average revenue based on the day of release
plt.figure(figsize=(10, 6))
sns.barplot(x=df['day'], y=df['revenue'], palette='Set1')
plt.title('Average Revenue by Release Day')
plt.xlabel('Release Day')
plt.ylabel('Average Revenue (in millions)')
plt.show()
```

* Highest revenue day: Tuesday (above 2.5e8 or 250 million).
* Other high performers: Wednesday and Monday.
* Lowest revenue day: Friday (below 1e8 or 100 million)

# 9.Average Revenue by Release Year

```python
# Barplot showing the average revenue based on the year of release
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['year'], y=df['revenue'], color="purple")
plt.title('Average Revenue by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average Revenue (in millions)')
plt.show()
```

* Overall trend: Average revenue has increased over time, especially after 2000.
* Fluctuations: There are notable spikes in revenue around 1980, 1990, and 2015.
* Recent peak: Around 2015, average revenue reaches its highest point in the data set (close to 3.5e8 or 350 million).

# 10.Top 10 Directors by Average Net Profit

```python
top_directors = df.groupby('director_name')['net_profit'].mean().sort_values(ascending=False).head(10)
top_directors
```

```python
# Barplot showing the average profit margin across directors

plt.figure(figsize=(10, 6))
sns.barplot( x=top_directors.index,y=top_directors.values, palette='Set3')
plt.title('Top 10 Directors by Average Net Profit')
plt.xlabel('Director')
plt.ylabel('Average Profit Margin (%)')
plt.xticks(rotation=45)
plt.show()
```

* James Cameron leads with the highest average net profit, significantly higher than the rest.
* Peter Jackson holds the second position, with roughly half the average net profit of James Cameron.
* Francis Lawrence is third, closely followed by Christopher Nolan in fourth place.
* Andrew Adamson and Michael Bay are nearly tied for fifth and sixth places, respectively.
* Jon Favreau, Joss Wan, Gore Verbinski, and Sam Mendes make up the lower four positions on the chart with relatively similar average net profits.

# Overall Insights:

### 1.Budget-Revenue Relationship:

Higher budgets often lead to higher revenues, evidenced by a strong positive correlation (0.71). However, some low-budget films still achieve significant revenues, indicating that budget isn't the only success factor.
Male-directed movies show more volatility in net profit, especially for higher-budget films, while female-directed movies show more stable profit trends.

### 2.Gender Impact:

Gender does not exhibit a clear influence on the budget-revenue relationship. While male-directed movies have a wider range of net profits and more outliers, female-directed movies tend to have more stable profitability.
The dataset contains more male-directed movies than female-directed ones.

### 3.Movie Popularity and Revenue:

Popularity is moderately correlated with revenue (0.59), indicating that more popular movies generally earn more. However, extreme outliers in revenue exist, regardless of popularity.
The number of votes also has a strong correlation with popularity (0.78), meaning movies with more votes tend to be more popular.

### 4.Profitability Trends:

The net profit shows a strong positive correlation with revenue (0.98), demonstrating that higher revenue almost always results in higher net profit.
Male directors have more extreme high-profit outliers, while female directors show less variability in net profits.

### 5.Seasonality and Revenue Trends:

Highest revenue months: May, June, and November, coinciding with holiday and summer blockbuster periods.
Lowest revenue months: January, September, and February.
Highest revenue days: Tuesdays, followed by Wednesdays and Mondays. Surprisingly, Friday, a traditional release day, shows the lowest average revenue.

### 6.Yearly Trends:

Revenue has increased significantly over time, especially after 2000, with notable spikes around 1980, 1990, and 2015. The highest revenue year is 2015.

### 7.Top Directors by Profitability:

James Cameron leads with the highest average net profit, significantly outperforming others.
Peter Jackson and Francis Lawrence also rank highly, while Christopher Nolan and Michael Bay have slightly lower but competitive profits.


# Suggestions and Recommendations:

###  1.Diversify Investments Across Budget Ranges:

While high-budget films generally yield higher revenues, investing in low-budget films with potential (such as indie movies) can still yield significant returns.

### 2.Female Director Representation:

Encourage more opportunities for female directors, as their movies show more stable profitability patterns, which could benefit long-term investments.

### 3.Leverage Popularity Metrics:

Increase marketing efforts for movies with moderate popularity to boost vote counts and engagement, which tend to lead to higher revenues.
Focus on creating more interactive campaigns (e.g., social media and fan engagement) to improve movie popularity before release.

### 4.Timing for Releases:

Schedule releases in high-performing months (May, June, and November) to maximize revenue potential.
Avoid releasing major films in low-revenue months (January, September, February) unless aiming for niche audiences or off-season awards.

### 5.Focus on Audience Engagement:

Increase focus on boosting vote count and user ratings by fostering positive word of mouth and user engagement after movie releases.
Use early screenings and critic reviews to build momentum before official release dates.

### 6.Director’s Influence:

Analyze and collaborate with top-performing directors, such as James Cameron, Peter Jackson, and Christopher Nolan, who have a proven track record of high profitability.
Consider the stable performance of female-directed movies for long-term planning, especially for mid-budget productions.
